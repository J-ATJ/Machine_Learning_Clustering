Es un algoritmo que se basa en la `densidad` de los datos, es decir funciona bajo la lógica de que elementos similares siempre están cerca unos de otros. La principal característica de `DBSCAN` es que solo asigna elementos a clusters cuando realmente pertenecen a él, de modo que "separa" aquellos elementos anómalos (ruido), contrario a otros algoritmos en los que siempre se asignan todos los datos.

Otra característica es que no se selecciona la cantidad de grupos a crear, estos se crean automáticamente por el modelo tomando en cuenta el `radio` y la cantidad de puntos vecinos `M` que se definen en las condiciones iniciales.

Problema: Clasificar a los clientes de una empresa según su "cercanía" unos con otros.

Librerías:

```python
import pandas as pd
import numpy as np 
from sklearn.cluster import DBSCAN 
```

DataFrame:

```python
df = pd.read_csv('./data.csv')
df.head()
```

Crear nArray:

```python
X = df.values
```

Crear modelo:

```python
model = DBSCAN(eps=3.8, min_samples=5)
```

Entrenar modelo:

```python
model.fit(X)
```

Obtener clasificaciones (-1 = elementos sin grupo):

```python
labels = model.labels_
set(labels)
```

Agregar columna de clasificaciones:

```python
df['labels'] = labels
df.head(2)
```

Librerías para graficar:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(font_scale=1.5)
sns.set_style("darkgrid")    
```

Gráfica 1:

```python
plt.figure(figsize=(10,6))

sns.scatterplot(data=df, x='edad', y='total_empleos', hue='labels', palette='bright', s=70)

plt.title("Total de empleos", fontdict={'fontsize': 20}, pad=15)
plt.show();
```

Gráfica 2:

```python
plt.figure(figsize=(10,6))

sns.scatterplot(data=df, x='edad', y='nivel_ingresos', hue='labels', palette='bright', s=70)

plt.title("Nivel de Ingresos", fontdict={'fontsize': 20}, pad=15)
plt.show();
```
